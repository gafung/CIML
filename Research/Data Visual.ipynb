```python
import pandas as pd
url = "data_v3.csv"
insider = pd.read_csv(url, header=0)
col_list = ['return_5d','side', 'return_t5', "return_t30", "vol_sh_out_pct", "stake_pct_chg", "tran_value","mkt_cap", "prev_tran_num","hit_rate_5d", "hit_rate_30d", "hit_rate_90d"]
insider = insider[col_list]
```

```python
import seaborn as sns; sns.set(style="ticks", color_codes=True)
g = sns.pairplot(insider, hue = 'return_5d', palette="husl", vars =['side', 'return_t5', "return_t30", "vol_sh_out_pct","stake_pct_chg", "tran_value","mkt_cap", "prev_tran_num","hit_rate_5d", "hit_rate_30d", "hit_rate_90d"])
import matplotlib.pyplot as plt
plt.show()
```
