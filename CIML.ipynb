```python
import os
from collections import namedtuple

StockFilepath = namedtuple('StockFilepath', ['stock', 'filepath'])
stock_filepath_pairs = []
for f in os.listdir(MARKET_DATA_FOLDER):
    fp = os.path.join(MARKET_DATA_FOLDER, f)
    if f.startswith('historical__') and f.endswith('.csv'):
        stock_str_start = len('historical__')
        stock_str_end = f.index(' HK ') - 1
        stock_str = f[stock_str_start: stock_str_end + 1]
        stock_filepath_pairs.append(StockFilepath(stock=stock_str, filepath=fp))
print(stock_filepath_pairs)
print(len(stock_filepath_pairs))
```

```python
def get_price_t_plus_n(stock_code, n, t, return_date):
    try:
        fp = next(x for x in stock_filepath_pairs if x.stock == stock_code).filepath
    except StopIteration:  # No market data for this stock
        return 'No Market Data'
    df_mrkt = pd.read_csv(fp, index_col=3, parse_dates=[3])  # index_col is date
    df_mrkt['date'] = df_mrkt.index.values
    try:
        offset = -1
        i = -1
        price = None
        current_date = t
        while i != n:  # while T+i != T+n
            i += 1
            offset += 1
            current_date = t+pd.DateOffset(offset)
            price = df_mrkt.at[current_date, 'PX_LAST']
            while pd.isnull(price):
                offset += 1
                current_date = t+pd.DateOffset(offset)
                price = df_mrkt.at[current_date, 'PX_LAST']
        return current_date if return_date else price
    except KeyError:  # No market data on this date
        return 'No Market Data'
    

df['T'] = df.apply(lambda sr: get_price_t_plus_n(sr['code'], 0, sr['date'], True), axis=1)
df['T+1'] = df.apply(lambda sr: get_price_t_plus_n(sr['code'], 1, sr['date'], True), axis=1)
df['PX_LAST T'] = df.apply(lambda sr: get_price_t_plus_n(sr['code'], 0, sr['date'], False), axis=1)
df['PX_LAST T+1'] = df.apply(lambda sr: get_price_t_plus_n(sr['code'], 1, sr['date'], False), axis=1)
df
```

```python
# Exclude those without market data
df2 = df[~(df['PX_LAST T+1']=='No Market Data')].copy()
df2
```
